SVM Model Training


```python
X = df_cancer.drop(['target'],axis = 1)
X
```

```python
y = df_cancer['target']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state=10)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
```

```python
svm =  SVC()
```

```python
svm.fit(X_train,y_train)
```

```python
y_pred = svm.predict(X_test)
y_pred
```

```python
from sklearn import metrics 
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
conf_mat = confusion_matrix(y_test, y_pred)
#sns.heatmap(conf_mat, annot = True)
conf_mat
```

Improving Model by Data Normalization (Feature Scaling)

1st method of Data Normalization using Min and Max Function

```python
#To Scale Train Data
min_train = X_train.min()
min_train
```

```python
range_train = (X_train-min_train).max()
range_train
```

```python
X_train_scale = (X_train - min_train)/range_train
X_train_scale
```

```python
#To Scale test Data
min_test = X_test.min()
range_test = (X_test-min_test).max()
X_test_Scale = (X_test-min_test)/range_test
```

2nd Method of Data Normalization by built in function


```python
#from sklearn.preprocessing import MinMaxScaler
#scaler = MinMaxScaler()
#X_train_scaled = scaler.fit_transform(X_train)
#X_test_scaled = scaler.transform(X_test)
```

```python
svm.fit(X_train_scale,y_train)
```

```python
y_pred =  svm.predict(X_test_Scale)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
conf_mat = confusion_matrix(y_test, y_pred)
#sns.heatmap(conf_mat, annot = True)
conf_mat
```

```python
print(classification_report(y_test, y_pred))
```
